```python
vid= cv2.VideoCapture(0)

while True:
    flag , img = vid.read()
    if flag:
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        cv2.imshow('Preview' ,img_grey)
        key = cv2.waitKey(1)
        if key == ord('q'):
            break
    else:
        break
    sleep(0.1)
    vid.release()
    cv2.destroyAllWindows()    
```

```python
vid= cv2.VideoCapture(0)

while True:
    flag , img = vid.read()
    if flag:
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        x1,y1,w,h=(200,400,200,200)
        img_cropped= img[y1:y1+h, x1:x1+w, :]
        
        cv2.imshow('Preview' ,img_cropped)
        key = cv2.waitKey(1)
        if key == ord('q'):
            break
    else:
        break
    sleep(0.1)
    vid.release()
    cv2.destroyAllWindows() 
```

```python
vid= cv2.VideoCapture(0)

while True:
    flag , img = vid.read()
    if flag:
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        x1,y1,w,h=(200,400,200,200)
        img_cropped= img[y1:y1+h, x1:x1+w, :]
        
        cv2.rectangle(
            img,
            plt1=(x1,y1), plt2=(x1+w, y1+h),
            color=(0,0,255),thickness=10
        )
        
        cv2.imshow('Preview' ,img)
        key = cv2.waitKey(1)
        if key == ord('q'):
            break
    else:
        break
    sleep(0.1)
    vid.release()
    cv2.destroyAllWindows() 
```
